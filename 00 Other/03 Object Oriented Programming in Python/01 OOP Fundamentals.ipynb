#### Standart Class

```python
class Employee:
    def set_name(self, new_name):
        self.name = new_name            # Setting the  name Attreibutes

    def set_salary(self,new_salary):
        self.salary = new_salary        # Setting the  salary Attreibutes

    def give_raise(self, amount):
        self.salary = self.salary + amount

    def monthly_salary(self):
        return self.salary / 12
```

```python
Employee1 = Employee()
Employee1.set_name('ujang')
Employee1.name
```

```python
Employee1.set_salary(10000)
Employee1.salary
```

```python
# Adding Value to the salary attributes

Employee1.salary = Employee1.salary+ 7999
Employee1.salary
```

```python
Employee1.monthly_salary()
```

```python
Employee1.give_raise(100000000)
print(Employee1.salary)
print(Employee1.monthly_salary())
```

# the __init attribute

The different is we dont need to call the method after creating the object, insted, we call it during the creation of the object (by using __init)

#### using 1 parameters

```python
class Customer:
    def __init__(self, name):
        self.name = name 
        print("the __init__ function is called for making name")
```

```python
cust1 = Customer() # Note that this object calling result an error, we need to put the parameter everytime init called
```

```python
cust1 = Customer('Ujang')
cust1.name
```

#### Using multiple parameters

```python
class Customer:
    def __init__(self, name, balance):
        self.name = name 
        self.balance = balance
        print("the __init__ function is called for making name")
```

```python
cust1 = Customer('Ujang', 20000)
print(cust1.name)
print(cust1.balance)
```

#### Setting default paramaeters value

```python
class Customer:
    def __init__(self, name, balance = 1000):
        self.name = name 
        self.balance = balance
        print("the __init__ function is called for making name")

cust1 = Customer('Ujang')
print(cust1.name)
print(cust1.balance)
```

# Best Practice

### Exercise


```python
import math

class Point:
    def __init__(self, x=0.0, y=0.0):
        self.x = x
        self.y = y

    def distance_to_origin(self):
        return math.sqrt(int(self.x)**2 + int(self.y)**2)

    def reflect(self, axis):
        
        
        if axis == 'x':
            self.y = - self.y
        elif axis == 'y':
            self.x = - self.x 
        else:
            print('Error')
```

```python
pt = Point(x=3.0)
pt.reflect("y")
print((pt.x, pt.y))
pt.y = 4.0
print(pt.distance_to_origin())
```
