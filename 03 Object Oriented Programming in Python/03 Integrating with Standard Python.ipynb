# Operator Overloading

This object is not equal : 





But, This custom object is considered as same



### Comparison Operators

to overcome that, we need to identify special method for that : __eq()

```python
class Customer:
    def __init__(self, id, name):
        self.id, self.name = id, name

    def __eq__(self, other) -> bool:  # Will be called when '==' is used
        print('__eq__ is called')
        return (self.id == other.id) and (self.name == other.name)
    

```

```python
A = Customer(1, 'Abid')
B = Customer(1, 'Abid')

print(A == B)
```

### Checking Class Equality

### Comparison of inheretance


```python
class Parent:
    def __eq__(self, other):
        print("Parent's __eq__() called")
        return True

class Child(Parent):
    def __eq__(self, other):
        print("Child's __eq__() called")
        return True
```

```python
p = Parent()
c = Child()

p == c 
```

```python
p = Parent()
c = Child()

c == p
```

# Operator Overloading: String representation

### Printing an Object

This printing object, result the memory information : 



While this print object result much information (the value of an object):



### Method for printing object representation

### Implementatation : __str

```python
class Customer:
    def __init__(self, name, balance):
        self.name, self.balance = name, balance

    def __str__(self):
        cust_string = ''' 
        Customer:
            name: {name}
            balance: {balance}
        '''.format(name = self.name, balance = self.balance)

        return cust_string
```

```python
Cust1 = Customer('Abid', 10000)
print(Cust1)  #Printing user friendly  output
```

```python
Cust1  # Will return the memory status
```

### Implementation : __repr

```python
class Customer:
    def __init__(self, name, balance):
        self.name, self.balance = name, balance

    def __repr__(self):
        return 'Customer({name}, {balance})'.format(name = self.name, balance = self.balance)
```

```python
Cust1 = Customer('Abid', 10000)
Cust1  #<<-- Will Implicity Call the __repr__
```

```python
print(Cust1)
```

### Exception Handling

### Raising Exceptions



```python
def make_list_of_ones(length):
    return [1]*length

make_list_of_ones(-2) # will return nothing
```

```python
def make_list_of_ones(length):
    if length <= 0:
        raise  ValueError("Invalid length!")   # Waill stop the program and raise an error
    return [1]*length
```

```python
make_list_of_ones(10)
```

```python
make_list_of_ones(-2) # will return the error
```

### Exceptions are classes and custom exceptions

```python
class BalanceError(Exception):
    pass    #Empty class is enaught to recognize it as an error class
```

```python
class Customer:
    def __init__(self, name, balance):
        if balance < 0:
            raise BalanceError("Balance should be on Positive Number!!!!!!")
        else:
            self.name, self.balance = name, balance
```

```python
Cust_a = Customer('Ujang', -10000)  # Will return an error
```

```python
Cust_a  # not that the object is not created at all
```

on user side, we can use try - except code :

```python
try:
    cust_a = Customer('Muhammad Abid', -100)
except BalanceError:
    cust_a = Customer('Muhammad Abid', 0)
```

```python
cust_a
```

```python
def invert_at_index(x, ind):
  try:
    return 1/x[ind]
  except ZeroDivisionError:
    print("Cannot divide by zero!")
  except IndexError:
    print("Index out of range!")
 
a = [5,6,0,7]

# Works okay
print(invert_at_index(a, 1))

```

```python
# Potential ZeroDivisionError
print(invert_at_index(a, 2))

```

```python
# Potential IndexError
print(invert_at_index(a, 5))
```
